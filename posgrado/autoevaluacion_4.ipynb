```python
import numpy as np

matrix_a = np.array([[1, 2, -1, 0], [2, 3, 1, 6], [0, 1, 0, 1]])

print(matrix_a)
```

```python
u, s, vh = np.linalg.svd(matrix_a)

print("U:\n", u)
print("S:\n", s)
print("Vh:\n", vh)
```

```python
a_daga = np.linalg.pinv(matrix_a)

print("A^+:\n", a_daga)
```

```python
x = np.array([1, 2, 3])

print(x.T)

print("A^+ * x:\n", a_daga @ x.T)
```
