```python
import numpy as np

matrix = numpy.array([[2/3, -1/3, -1/3], [-1/3, 2/3, -1/3], [-1/3, -1/3, 2/3]])

print(matrix)
```

```python
idempotent = matrix @ matrix
print(idempotent)
```

```python
if (np.allclose(matrix, idempotent)):
    print("La matriz es idempotente")
```

```python
a = np.array([[1, 3 ,2], [4, 2, -1], [5, 5, 1]])

print(a)
```

```python
autovalores = np.linalg.eigvals(a)
print(np.round(autovalores))
```

```python
print(a @ np.array([5/3, 1, 8/3]))
print(a @ np.array([1/np.sqrt(2), -1/np.sqrt(2), 0]))
print(a @ np.array([-1, 1, 0]))
```

```python
b = np.array([[1, 3 ,2], [4, 2, -1], [5, 5, 1]])
identidad = np.eye(3)*6
print(identidad)

rango = np.linalg.matrix_rank(b - identidad)

multiplicidad = len(b) - rango
print(multiplicidad)
```
